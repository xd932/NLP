```python
# Bag of Words
# USes cosine for vector similarity
#order of words, context not matter
#matter on frequency of words
# words which are not in vocalbulary are taken care

df=pd.DataFrame({'text':["The brown fox jumps over the lazy dog.",
"A quick brown dog jumps over the lazy fox.",
"The lazy dog sleeps all day.",
"A lazy fox naps in the sun."],'output':[1,1,0,0]})
print("DF: ", df)


from sklearn.feature_extraction.text import  CountVectorizer
cv=CountVectorizer()
bag=cv.fit_transform(df['text'])

```

```python
print(cv.vocabulary_)
print(bag[0].toarray())
print(bag[1].toarray())
```

```python
#Tf-Idf
# Assign each words value(weightage)
# TF(Term frequency)
# IDF(inverse document frequency)


# d1="The brown fox jumps over the lazy dog.",
# d2="A quick brown dog jumps over the lazy fox.",
# d3="The lazy dog sleeps all day.",
# d4="A lazy fox naps in the sun."



# TF(t,d)=(no of occurence of term(t) in doccument (d))/(no of terms in document(d))
# Also between [0,1]
# Ex: TF(brown,d1)=(1/8)

#IDF(t)=Log[(No of documents(d) in corpus)/(no of document(d) with term (t) in them)]
#IDF(dog)=Log(4/3)


# Need log in idf Because it need to multiple with TF, if log not usede IDf , it will dominate since it will have large value 
# TF*IDF


# IMp: in skelarn actual 

# #IDF(t)=Log[(No of documents(d) in corpus)/(no of document(d) with term (t) in them)]+1
# to avoid ignoring words with idf '0'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
