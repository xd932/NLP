```python
#train your own data 
import os 
import gensim
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
books = ['/kaggle/input/game-of-thrones-books/001ssb.txt',
         '/kaggle/input/game-of-thrones-books/002ssb.txt',
         '/kaggle/input/game-of-thrones-books/003ssb.txt',
         '/kaggle/input/game-of-thrones-books/004ssb.txt',
         '/kaggle/input/game-of-thrones-books/005ssb.txt']

for file_path in books:
    with open(file_path, 'r', encoding='latin-1') as file:
        content = file.read()
        raw_sent = sent_tokenize(content)
        for sentence in raw_sent:
            story.append(simple_preprocess(sentence))

```

```python
len(story)
```

```python
story[:2]
```

```python
model=gensim.models.Word2Vec(window=10, min_count=2)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
#default=5
```

```python
model.wv.most_similar('angry')
```

```python
model.wv.doesnt_match(['jon','robb','arya','bran','rikon','sansa'])
```
